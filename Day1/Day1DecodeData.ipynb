```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
y1 = x ** 2
y2 = 2 * x + 3
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
themes = plt.style.available
print(themes)
```

```python
plt.plot(x,y1, marker='*')
plt.plot(x,y2, marker='o', markersize=12, linestyle='dashed')
plt.show()
```

```python
plt.style.use('seaborn')
```

```python
plt.plot(x,y1, marker='*', label='Iphone')
plt.plot(x,y2, marker='o', markersize=12, linestyle='dashed', label='Android')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title("IEEE CIS")
plt.legend()
plt.show()
```
